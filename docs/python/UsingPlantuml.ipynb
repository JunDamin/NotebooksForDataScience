## 왜 plantUML인가?

변동사항을 관리하기가 편하다.
이름을 지정할 수 있어서 덩어리로 관리할 수 있으며 변화를 관리하기가 편하다.
아주 세밀하게 수정할 것이 아니라면, 빠르게 그럭저럭 쓸만한 그림을 만들어 낼 수 있다.

내가 원하는대로 배치하는 건 거의 불가능하다고 볼 수 있다.

## 세팅하기

여러가지 방법이 있겠지만, 파이썬과 jupyter notebook을 이용하는 사람에게는 `iplantuml`을 사용하는 것이 가장 간편하다.

## 기본원리

가명 사용 => 거꾸로 해도 가능하다

컴포넌트 클래스 등등 기존에 사용되는 것이 있다.

여러가지 다른 곳에서 사용되던 것이 있어서 익숙하지 않을 수는 있다.

화살표의 종류가 다양하다. 대부분은 --> 또는 --|> 정도만 쓸 것이다. 

중괄호를 사용해서 메소드를 설정할 수 있다.
또한 메소드에 마커를 넣어서 표현 할 수도 있다.
클래스 안에 -- (단어) -- 나 ..(단어).., ==(단어)== 등을 통해서 선을 그을 수 있다.

<<(L, #FFEEAA) name>> 등을 통해서 Spot의 단어를 바꿀 수 있다.


package 중괄호 안에 작성해서 묶을 수 있다.

namespace 를 통해서 같음 명칭의 클래스를 쓸 수 있다.

left right direction 을 통해서 방향을 조정할 수 있다.

테마를 정해서 조정할 수 있다.



```python
import iplantuml
```

```python
%%plantuml

@startuml
!theme mars

title "As Is"

card 이동대상 {
(부서 및 직무 정보 취합)
(개인 동기 및 능력)
(직무희망조사서 작성)    
(정보 분석 및 검토)
}

card 인사부서 {
(직무희망조사 취합)
(부서인력 수요 취합)
(검토 및 의사결정) 
(부서배치)    
}

(개인 동기 및 능력) --> (정보 분석 및 검토)
(부서 및 직무 정보 취합) ==> (정보 분석 및 검토) #line:blue : "4. 소극적 현황 파악 노력"
(정보 분석 및 검토) ==> (직무희망조사서 작성) #line:blue : "5. 낮은 체감효용으로 소극적 작성"

(직무희망조사서 작성) ==> (직무희망조사 취합) #line:blue : "1. 희망부서 및 직무 정보만 취합" 
(직무희망조사 취합) ==> (검토 및 의사결정) #line:blue
(부서인력 수요 취합) --> (검토 및 의사결정)
(검토 및 의사결정) ==> (부서배치) #line:blue : "2. 극소수의 인원만 희망부서 배치"
(부서배치) ..> (부서 및 직무 정보 취합) #line:blue : "3. 직무희망조사에 대한 체감효용 저하"
(부서배치) ==> (업무성과) #line:blue : "낮은 직무만족도로 인한 성과저하"


@enduml
```

```python
%%plantuml

@startuml
!theme mars

title "To Be"

card 이동대상 {
(부서 및 직무 정보 취합)
(개인 동기 및 능력)
(직무희망조사서 작성)    
(정보 분석 및 검토)
}

card 인사부서 {
(직무희망조사 취합)
(부서인력 수요 취합)
(검토 및 의사결정) 
(부서배치)    
}

(개인 동기 및 능력) --> (정보 분석 및 검토)
(부서 및 직무 정보 취합) ==> (정보 분석 및 검토) #line:red : "4. 적극적 현황 파악 노력"
(정보 분석 및 검토) ==> (직무희망조사서 작성) #line:red : "5. 높은 체감효용으로 적극적 참여"

(직무희망조사서 작성) ==> (직무희망조사 취합) #line:red : "1. 비희망 부서 및 직무 정보까지 취합"

(직무희망조사 취합) ==> (검토 및 의사결정)  #line:red
(부서인력 수요 취합) --> (검토 및 의사결정)
(검토 및 의사결정) ==> (부서배치) #line:red : "2. 비희망직무 배치 최소화"
(부서배치) ..> (부서 및 직무 정보 취합) #line:red : "3. 직무희망조사에 대한 체감효용 제고"
(부서배치) ==> (업무성과) #line:red : "개선된 직무만족도로 인한 성과제고"


@enduml
```
