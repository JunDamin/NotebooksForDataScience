# 1 서문

많은 프로그래밍 언어가 있으며 각 언어는 각각의 장점과 약점을 가지고 있습니다. 어떤 언어는 아주 빠르지만 장황합니다. 또 다른 언어는 아주 쓰기 쉽지만, 느립니다. 이 문제는 *두 언어* 문제로 알려져 있으며 줄리아는 이 문제를 우회하는 것을 목표로 하고 있습니다. 비록 우리 셋 모두는 다른 분야에서 일하고 있지만, 우리 모두는 우리가 사용했던 다른 언어보다 줄리아가 우리 연구에 좀더 효과적임을 발견했습니다. 우리는 이에 대한 논의를 섹션2에 하고 있습니다. 그러나 다른 언어에 비해, 줄리아는 새로운 언어입니다. 이것은 언어를 둘러싼 생태계를 여행는데 때때로 어렵다는 뜻입니다. 여러 패키지들이 어떻게 합쳐지는지 알아내는 건 어렵습니다. 이것이 우리가 이 책을 쓰기로 한 이유입니다! 우리는 이 놀랍고 새로운 언어를 시작하는 것이 쉬워질 수 있기를 바랬습니다. 특히 우리 동료들에게 말입니다.

위에서 논의한 것처럼 각 언어는 자신의 강점과 약점이 있습니다. 우리는 데이터 사이언스가 줄리아의 강점 이라고 생각합니다. 동시에 우리 세 모두 데이터 사이언스 도구를 우리의 매일 일상에 사용하고 있습니다. 그리고 아마도 당신도 그럴 겁니다! 그렇기 때문에 이 책은 데이터 사이언스에 집중하고 있습ㄴ다.

다음 파트에서 우리는 **데이터 사이언스에서 데이터** 부분을 강조하고자 하며 왜 데이터 스킬이 학계뿐만 아니라 업계에 **높은 수요**가 있고 그렇게 유지될 것을 강조하고자 합니다. 우리는 동료들에게 코드를 공유하고 갱신 할 때의 마찰을 줄여줄 **소프트웨어 엔지니어링 연습과 결합된 데이터 사이언스**에 대해서 논의할 것입니다. 
모든 데이터 분석은 협력적인 노력입니다. 그렇기 때문에 이런 소프트웨연습이 당신을 도울 것입니다.

### 1.0.1 데이터는 어디에나 있다

**데이터는 풍부합니다**. 그리고 멀지 않은 미래에 더 많아질 것입니다. 2012년 발표된 한 보고서에서 2005년부터 2020년까지 디지털화 되어 보관된 데이터가 **300배, 130엑사바이트에서 터무니 없이 큰 40,000엑사바이트**로 증가될 것이라고 결론 내렸습니다.([Gantz&Reinsel, 2012]()) 이것은 40조 기가바이트로, **지금 지구상에 존재하는 인류 1인당 5.2 테라바이트**에 해당하는 양입니다.
2020년에는, 평균적으로 모든 사람들이 **초당 1.7MB**의 자료를 생산하다고 합니다. ([Domo, 2018]()). 최근 보고서에서는 **국가 GDP의 65%가 디지털화 될 것**이라고 보고 있습니다.([Fitzgerald et al., 2020]()).

모든 직업은 늘어나는 데이터 가용성과 데이터의 중요성에 영향을 받게 될 것입니다.([Chen et al., 2014](); [Khan et al., 2014]()). 데이터는 통신과 지식 형성, 의사 결정에 사용됩니다. 이것이 데이터 스킬이 중요한 이유입니다. 당신이 데이터 처리에 익숙하다면, 당신은 가치 있는 연구자나 프로가 될 수 있습니다. 다른 말로 당신은 **데이터 문해력**이 있어집니다.

## 1.1 데이터 과학이란 무엇인가?

데이터 과학이란 머신러닝과 통계만이 아니고, 또한 예측 뿐이 아닙니다. 또한, STEM(과학, 기술, 공학, 수학) 분야에 완전히 포함되는 훈련도 아닙니다. 하지만 우리가 한가지 확실히 말할 수 있는 건 데이터 과학은 언제나 **데이터**에 관한 것입니다. 우리의 이책의 목표는 두부분으로 정리 됩니다.

- 우리는 데이터 과학의 중추에 집중한다 : **데이터**
- 우리는 **줄리아** 프로그래밍 언어로 데이터를 처리한다

우리는 왜 줄리아가 데이터 과학에 있어 극도로 효과적이 언어인지 섹션 2에서 논의할 예정입니다. 지금은 우리의 관심을 데이터로 옮겨 봅시다.

### 1.1.1 데이터 문해

[위키피디아]()에 의하면, **데이터 문해란 "데이터를 정보로 읽고, 이해하며, 생산하고 의사소통할 수 있는 능력"**이라고 정의하고 있습니다. 데이터 문해력이 있다는 것은 당신이 데이터에 압도된다는 느낌을 받지 않고, 오히려 이를 사용해서 올바른 의사결정을 한다는 좀더 편안한 아이디어도 우리는 좋아합니다. 데이터 문해력은 아주 경쟁력 있는 기술이라고 볼 수 있습니다. 이 책에서 우리는 데이터 문해의 두 측면을 커버합니다.

1. `DataFrames.jl`을 사용한 **데이터 조작**(섹션 [4]()). 이 쳅터에서 당신은 어떻게 아래와 같은 것을 할 수 있는지 배웁니다.
  1. CSV와 Excel 데이터를 줄리아로 읽어오기
  2. 데이터 질문에 답하도록 줄리아에서 데이터 처리
  3. 데이터 필터링과 부분 집합 구하기
  4. 누락된 데이터 처리하기
  5. 여러 데이터 소스를 합치기
  6. 그룹화하고 데이터 요약하기
  7. CSV와 Excel파일로 줄리아에서 데이터 내보내기
2.`Makie.jl` 사용한 **데이터 시각화**(섹션 [5]()). 이 쳅터에서 당신은 어떻게 아래와 같은 것을 하는 것을 배웁니다.
  1. `Makie.jl`의 다양한 백엔드를 사용해서 데이터 플롯 그리기
  2. 시각화된 자료를 PNG나 PDF와 같은 여러 포맷으로 저장하기
  3. 다양한 데이터 시각화를 여러 플롯 함수를사용해서 만들기
  4. 속성들을 통해 시각화 조정하기
  5. 새로운 테마를 만들고 적용하기
  6. $\LaTeX$ 요소를 플롯에 넣기
  7. 컬러와 팔레트 조작하기
  8. 복잡한 레이아웃 만들기

## 1.2 소프트웨어 엔지니어링

다른 대다수의 데이터 과학 책과 달리, 이 책은 적절히 **구조화된 코드**에 좀더 강조점을 두고 있습니다. 우리는 많은 데이터 사이언티스트가 그들의 코드를 한가지 큰 파일에 넣고 모든 선언문을 순서대로 실행시킨다는 것을 알게 되었기 때문입니다. 이것은 독자에게 언제나 처음부터 끝까지 읽어야 하고 이전 섹션으로 다시 돌아가거나, 재미 있는 섹션으로 바로 넘어가는 것을 하지 못하도록 하는 것과 같다고 생각할 수 있습니다.
이런 방식은 작고 간단한 프로젝트에서는 괜찮습니다. 하지만 프로젝트가 커지고 복잡해질 수록 더 많은 문제들이 발생하기 시작합니다. 예를 들어, 잘 쓰여진 책은 명확히 구분된 이름으로 된 챕터와 섹션으로 나뉘어져 있고 다른 부분을 인용이 포함되어 있습니다. 소프트웨어에서 이와 동등한 것이 바로 **코드를 함수로 나누어 두는 것**입니다. 각 함수는 이름과 내용이 있습니다. 함수를 사용해서, 당신은 컴퓨터에게 당신의 코드에서 다른 곳으로 넘어가서 이어서 실행되게 할 수 있습니다. 이것은 여러 프로젝트에서 코드를 재사용, 코드 갱신, 코드 공유, 협업 그리고 큰 그림을 보는 것을 좀더 쉽게 해줍니다.따라서 함수를 쓰면 당신은 **시간을 아낄 수 있습니다.**

그래서 이 책을 읽는 동안, 당신은 결국 함수를 읽고 쓰는 것에 익숙해 질 겁니다. 
좋은 소프트웨어 엔지니어링 스킬이 주는 또 다른 장점은 좀더 쉽게 소스코드를 읽을 수 있게 해준다는 것입니다. 이런 장점은 당신의 코드를 디버깅할 때 정말 큰 도움이 되며 어떻게 패키지가 돌아가는지 이해하는데도 도움을 줍니다.마지막으로 말씀드리고 싶은 것은 우리가 이런 강조를 시작한게 아니라는 점입니다. 현업에서 개발자에게 "**주석 대신 함수를 쓰도록** 격려합니다. 이것은 사람에게 주석을 쓰고 컴퓨터를 위해 코드를 쓰기 보다, 개발자가 함수를 써서 사람과 컴퓨터에게 모두 보이도록 하는 것입니다.

또한 우리는 일관된 스타일을 유지하기 위해 많은 노력을 했습니다. 프로그래밍 스타일 가이드는 코드를 쓸 때 가이드라인을 줍니다. 예를 들어 공백을 어디에 넣는지, 어떤 이름이 대문자여야 하는지 말입니다. 엄격히 스타일을 지키는 것은 현학적으로 보일지도 모르겠습니다. 실제로도 그럴 때가 있습니다. 하지만, 코드가 일관될수록, 읽 이해하는 것이 쉬워집니다. 우리 코드를 읽을 때 우리 스타일 가이드를 알 필요는 없습니다. 당신은 읽으면서 알게 될 것입니다. 당신이 좀더 자세한 스타일 가이드를 원한다면 섹션 6.2를 확인해 주세요.
